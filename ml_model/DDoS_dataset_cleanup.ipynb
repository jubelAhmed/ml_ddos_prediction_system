```python
original_path = '/content/drive/My Drive/cic_ddos_2019/original'
destination_path = '/content/drive/My Drive/cic_ddos_2019/parquet_files/'
```

```python
dataframes = []
filepaths = []
for root, dirs, files in os.walk(original_path):
  for file in files:
    if file.endswith('.csv'):
      file_path = os.path.join(root, file)
      filepaths.append(file_path)
      df = pd.read_csv(file_path)
      dataframes.append(df)

# Access the individual dataframes
for df in dataframes:
  print(df.head())
```

```python
def clean_column_names(df):
  df.columns = df.columns.str.strip().str.replace('/', '_').str.replace(' ', '_').str.lower()
  return df
```

```python
dataframes = [clean_column_names(df) for df in dataframes]
```

```python
def check_duplicates_nan(df):
    print("\nDuplicate and NaN values")
    num_duplicates = df.duplicated().sum()
    print(f"Duplicates: {num_duplicates}")

    # Print the count of NaN values for each column, sorted in descending order
    nan_counts = df.isna().sum().sort_values(ascending=False)
    print("NaN values per column (highest first):")
    print(nan_counts)

for df in dataframes:
  check_duplicates_nan(df)
```

```python
def remove_duplicates_nan(df):
  df.dropna(inplace=True)
  df.drop_duplicates(inplace=True)
  return df

dataframes = [remove_duplicates_nan(df) for df in dataframes]

# Verify that duplicates and NaN values have been removed
for df in dataframes:
  check_duplicates_nan(df)
```

```python
def downsize_dtypes(df):
  for col in df.columns:
    if df[col].dtype == 'int64':
      df[col] = pd.to_numeric(df[col], downcast='integer')
    elif df[col].dtype == 'float64':
      df[col] = pd.to_numeric(df[col], downcast='float')
  return df

updated_dataframes = [downsize_dtypes(df) for df in dataframes]
```

```python
import os


if not os.path.exists(destination_path):
  os.makedirs(destination_path)
  print(f"Directory '{destination_path}' created successfully.")
else:
  print(f"Directory '{destination_path}' already exists.")


for i, df in enumerate(updated_dataframes):
  # df['simillarhttp'] = df['simillarhttp'].astype(str)
  file_name = f"{filepaths[i].split('/')[-1].replace('.csv', '.parquet')}"
  print(file_name)
  output_path = os.path.join(destination_path, file_name)
  df.to_parquet(output_path, index=False)  # Set index=False to avoid saving the index

print("Parquet files saved successfully!")
```

```python
import os

def get_file_sizes_in_mb(destination_path):
  """Prints the sizes of files in the given directory in MB."""

  for filename in os.listdir(destination_path):
    filepath = os.path.join(destination_path, filename)
    if os.path.isfile(filepath):
      size_in_bytes = os.path.getsize(filepath)
      size_in_mb = size_in_bytes / (1024 * 1024)  # Convert bytes to MB
      print(f"{filename}: {size_in_mb:.2f} MB")


```
